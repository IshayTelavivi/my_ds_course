## Rock Scissors Paper Game

### The purpose of the game is to win the most rounds.

At the beginning the players are asked to write their names.

Each round the players are asked to type Rock / Scissors / Paper, and acoording to their choice , one of them wins. If the match comes up in a tie, the players are given another try, so that there must be a winner for each round. If the player types an error, the program asks to retype.

At the end of each round the players are asked whether they want to continue playing. Only if both want, the game continues.

When the game ends, the program privides a game summary.

The code is composed of three functions:
* one_round - the function generates one round of the game. The function takes no input (the input is within the function), and provides output as the winner's name and the guesses of both players.

* count_games - the function counts the number of winning rounds, provides an intermediate score message and checks if the plaers want to continue. The output is number of rounds, the game winner, the guesses for each round and the intermediate score for each round.

* show_game_summary - the function provides a game summary with three parts: winner of the whole game message, a table with rounds details, and a summary of numbers of wins per player.

I did not write a funtion to the names submission, to avoid a situation where an input is required whenever the function is called.

I combined the whole code to one cell, so that it runs in one pulse. 

```python
# Below are dictionaries to translate guesses as string(player's guess) to numbers and vise versa
GUESS_TO_NUMBER = {"Rock": 1, "Scissors": 2, "Paper": 3} # Turns player's guess into an integer to allow calculating result
NUMBER_TO_GUESS = {1: "Rock", 2: "Scissors", 3: "Paper"} # Turns computer's random chiose into string to be printed as a gues

names = []
def get_names():
    
    # First we ask for the players' names
    player_1_name = input("Player 1, please write your name:")
    print("Welcome " + player_1_name + "!\n")
    names.append(player_1_name)
    # I use the following loop to make sure that the names are different, so that there is no mix up during the game
    player_2_name = player_1_name
    while player_2_name == player_1_name:
        player_2_name = input("Player 2, please write your name:")
        if player_2_name == player_1_name:
            print("Names Should be different. Please provide a different name.\n") # A message if player 2 provides the same name
    names.append(player_2_name)
    print("Welcome " + player_2_name + "!\n")
    print("Let's start playing.\n")


def one_round():
    
    result = 0 # This is a variable that determines who wins. While round comes out as tie (result=o), it will reiterate
    while result == 0: # Using a while loop, to allow another round if game comes out in a tie
        guess_numbers = [] # List to collect the choices of the players
        for i in range(2): # For each player
            player_guess = None
            while player_guess not in GUESS_TO_NUMBER: # This while loop is for the case of type error
                player_guess = input(names[i] + ", choose Rock, Scissors or Paper:")
                if player_guess not in GUESS_TO_NUMBER:
                    print(names[i] + ", you typed an error. Please type correctly one of the three options.\n") 
                    continue
                player_number = GUESS_TO_NUMBER[player_guess] # Translate guess to numbers, to allow winning result calulation
                guess_numbers.append(player_number)
 
    
        # At this point we have an integer (1, 2 or 3) for both players.

        #The following boolean expression deternmined the winner
        result = guess_numbers[0] - guess_numbers[1]
        # If results is 1 or -2 player 2 wins. If results is -1 or 2 player 1 wins
        if result == 0: # a tie
            print("Round came out in a tie. Please try again.\n")
        elif result == 1 or result == -2:
            round_winner = names[0] # Player 1
        else:
            round_winner = names[1] # Player 2
    print("{}, you win this round.\n".format(round_winner)) # Messgae at the end of each round
    return (round_winner, NUMBER_TO_GUESS[guess_numbers[0]], NUMBER_TO_GUESS[guess_numbers[1]])



def game_iterate_and_count():
    
    game_counter = 0 # Count number of rounds
    player_1_score = 0 # Count wins of player 1
    player_2_score = 0 # count wins of player 2
    game_result_list = [] # Collect winner per round
    intermediate_score_list = [] # Collects imtermediate scores
    is_want_play = True
    
    while is_want_play: # If at least one of the players doesn't want to continue playing, the loop ends
        game_counter += 1
        current_round = one_round() # 'current_round' is a tuple with the round result (round winner, player 1 guess, player 2 guess)
        round_winner = current_round[0]
        # The following line keeps the round result in the dictionary
        game_result_list.append([round_winner, current_round[1], current_round[2]]) 
        # The following boolean expression adds a point to the winner
        if round_winner == names[0]:
            player_1_score += 1
        else:
            player_2_score += 1
        # The following 3 lines summarize intermediate score, print it, and collect it to a dictionary
        intermediate_score = [player_1_score, player_2_score]
        print("Score is: {} - {} wins. {} - {} wins".format(names[0], player_1_score, names[1], player_2_score))
        intermediate_score_list.append(intermediate_score)
        print("Do you want to continue to the next round?")
        cont_1_input, cont_2_input = None, None
        # The following while verifies that the input is valid (Y/N)
        while (cont_1_input != 'Y' and cont_1_input != 'N') or (cont_2_input != 'Y' and cont_2_input != 'N'):
            cont_1_input = input(names[0] + " (Y/N):")
            cont_2_input = input(names[1] + " (Y/N):")
            print()
            if (cont_1_input != 'Y' and cont_1_input != 'N') or (cont_2_input != 'Y' and cont_2_input != 'N'):
                print("Answer should be 'Y or 'N'.")
        # This is the condition for continuing the game
        if cont_1_input == 'N' or cont_2_input == 'N':
            is_want_play = False
    # The next condition determines the game winner
    if player_1_score > player_2_score:
        game_winner = names[0] # Player 1
    elif player_2_score > player_1_score:
        game_winner = names[1] # Player 2
    else:
        game_winner = "No one" 
    return (game_counter, game_winner, game_result_list, intermediate_score_list)


def show_game_summary():
    
    game_summry = game_iterate_and_count() # Brings the information for all rounds
    print("\nGame Summary:\n\nThe winner of the game is {} !!!\n".format(game_summry[1]))
    row = "{:15} {:20} {:25} {:25} {:^10}" # This allows arranging the summary in columns
    print(row.format("Round #", "Round Winner", player_1_name + " Guess", player_2_name + " Guess", "Round Score"))
    print("-" * 100)
    for round_number in range(game_summry[0]): # Every round gets a row
        round_num = 'Round-' + str(round_number+1)
        round_winner = game_summry[2][round_number][0]
        player_1_guess = game_summry[2][round_number][1]
        player_2_guess = game_summry[2][round_number][2]
        round_score = str(game_summry[3][round_number][0]) + " : " + str(game_summry[3][round_number][1]) 
        print(row.format(round_num, round_winner, player_1_guess, player_2_guess, round_score))
    # The for loop below gives the number of wins per player
    for i in range(2):
        player_wins = game_summry[3][game_summry[0]-1][i] # Within the intermediate_score_list, go to last game, and to relevant player
        print("{} won {} rounds.".format(names[i], player_wins))

get_names()   
show_game_summary()
```
